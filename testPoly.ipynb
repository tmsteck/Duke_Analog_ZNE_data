```python
import numpy as np
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt


test_funciton = lambda x: 1 + 2*x**2 + 3*x**3 + 4*x**4 + 5*x**5
test_x = np.linspace(-1, 1, 100)
plt.plot(test_x, test_funciton(test_x), label='test function')
polynomial_fit_func = Polynomial.fit(test_x, test_funciton(test_x), [0,2,3,4,5])
plt.plot(test_x, polynomial_fit_func(test_x), label='polynomial fit')
print(polynomial_fit_func)
```
